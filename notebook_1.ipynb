```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from pathlib import Path

np.random.seed(42)
```

```python
datapath = Path() / "data" / "lifesat"

oecd_bli = pd.read_csv(datapath / "oecd_bli.csv")
gdp_per_capita = pd.read_csv(datapath / "gdp_per_capita.csv")
```

```python
display(oecd_bli)
```

```python
display(gdp_per_capita)
```

```python
gdp_per_capita_2020 = gdp_per_capita[gdp_per_capita["Year"]==2020]
gdp_per_capita_2020 = gdp_per_capita_2020.rename(
    columns={"GDP per capita, PPP (constant 2017 international $)": "GDP per capita $"})
gdp_per_capita_2020
```

```python
oecd_bli["Indicator"].value_counts()
```

```python
oecd_bli = oecd_bli[oecd_bli["INEQUALITY"]=="TOT"]
oecd_bli = oecd_bli.pivot(index="Country", columns="Indicator", values="Value")
oecd_bli
```

```python
oecd_bli = oecd_bli.reset_index()
oecd_bli
```

```python
full_country_stats = oecd_bli.merge(gdp_per_capita_2020, left_on="Country", right_on="Entity", how="inner")
full_country_stats
```

```python
full_country_stats = full_country_stats[["Country","Life satisfaction", "GDP per capita $"]]
full_country_stats = full_country_stats.sort_values(by="GDP per capita $")
full_country_stats
```

```python
full_country_stats.to_csv(datapath / "lifesat_full.csv", index=False)
```

```python
min_gdp = 23_500
max_gdp = 62_500

country_stats = full_country_stats[(full_country_stats["GDP per capita $"] >= min_gdp) &
                                    (full_country_stats["GDP per capita $"] <= max_gdp)]
country_stats.to_csv(datapath / "lifesat.csv", index=False)
```

```python
lifesat = pd.read_csv(datapath / "lifesat.csv")

lifesat.rename(columns={"GDP per capita $": "PKB na mieszkanca $"}, inplace=True)
lifesat.rename(columns={"Life satisfaction": "Satysfakcja z Å¼ycia"}, inplace=True)

lifesat
```

```python
lifesat.plot(kind="scatter", grid=True, x="PKB na mieszkanca $", y="Satysfakcja z Å¼ycia")
plt.show()
```

```python
x = lifesat[["PKB na mieszkanca $"]]
y = lifesat[["Satysfakcja z Å¼ycia"]]

model = LinearRegression()

model.fit(x, y)

theta_0 = model.intercept_[0]
theta_1 = model.coef_[0][0]

print(f"theta_0 {theta_0}")
print(f"theta_1 {theta_1}")
```

```python
lifesat.plot(kind="scatter", grid=True, x="PKB na mieszkanca $", y="Satysfakcja z Å¼ycia")

plt.plot(x, theta_0+theta_1*x)
plt.text(45000, 6.5, fr"$\theta_0 = {theta_0.round(2)}$", color="blue")
plt.text(45000, 6.3, fr"$\theta_1 = {theta_1 * 1e5:.2f}\cdot 10^{{-5}}$", color="blue")

plt.show()
```

```python
cyprus_gdp_per_capita_2020 = gdp_per_capita_2020[gdp_per_capita_2020["Entity"] == "Cyprus"]
cyprus_gdp_per_capita_2020 = cyprus_gdp_per_capita_2020[["GDP per capita $"]]
cyprus_gdp_per_capita_2020.rename(columns={"GDP per capita $": "PKB na mieszkanca $"}, inplace=True)
cyprus_gdp_per_capita_2020
```

```python
cyprus_predicted_life_satisfaction = model.predict(cyprus_gdp_per_capita_2020)
cyprus_predicted_life_satisfaction
```

```python
from sklearn.neighbors import KNeighborsRegressor
model_knn = KNeighborsRegressor(n_neighbors=3)

model_knn.fit(x, y)

model_knn.predict(cyprus_gdp_per_capita_2020)
```
